Name: Online Shoppers Intention Dataset

Description: Contains information about online shoppers' behavior and purchase outcomes.


Features:

Administrative, Informational, ProductRelated: Number of pages visited in each category.

Administrative_Duration, Informational_Duration, ProductRelated_Duration: Time spent on each category's pages.
BounceRates, ExitRates, PageValues: Website engagement metrics.

SpecialDay, Month, OperatingSystems, Browser, Region, TrafficType, VisitorType, Weekend: Context of the visit.


Revenue: Indicates whether a purchase was made (target variable).

### Problem Statement:

Goal: Predict whether an online shopper will make a purchase based on their browsing behavior and session characteristics.




Type of Problem: Binary classification (predicting a "Yes" or "No" outcome for purchase).

# 3.3 handling the outliers

```python


# Function to handle outliers using IQR method
def handle_outliers_iqr(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)

# Apply the handle_outliers_iqr function to each numerical column
for column in numerical_columns:
    handle_outliers_iqr(column)

# Box plots after handling outliers
plt.figure(figsize=(12, 6))
df[numerical_columns].boxplot()
plt.xticks(rotation=45)
plt.title("Box plots of Numerical Variables (After Handling Outliers)")
plt.show()
```

### 5.1 creating new features based on current features 

```python
import pandas as pd


# Session Duration
df['Session_Duration'] = df['Administrative_Duration'] + df['Informational_Duration'] + df['ProductRelated_Duration']

# Interaction Rate
df['Interaction_Rate'] = (df['Administrative'] + df['Informational'] + df['ProductRelated']) / (1 + df['Session_Duration'])

# Page Value per Product
df['PageValue_Per_Product'] = df['PageValues'] / (1 + df['ProductRelated'])

# SpecialDay Binary Indicator
df['SpecialDay_Indicator'] = (df['SpecialDay'] > 0).astype(int)

# Weekend and Revenue Interaction
df['Weekend_Revenue_Interaction'] = df['Weekend'] & df['Revenue']

# Bounce and Exit Rates Interaction
df['Avg_Bounce_Exit_Rates'] = (df['BounceRates'] + df['ExitRates']) / 2

# Display the updated DataFrame
print(df.head())
```
